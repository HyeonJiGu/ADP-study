# 정상성

```python
import statsmodels.api as sm
fig = plt.figure(figsize = (12,7))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts, lags = 20, ax = ax1)
```

acf값이 아주아주 천천히 작이지는 것을 볼 수 있으며, 이는 정상성을 만족하지 않는다는 것을 알 수 있다.

##### 단위근 검정인 ADF 검증을 통한 정상성 확인
- H0(귀무가설) : 자료에 단위근이 존재한다. 즉 정상성을 만족하지 않는다.
- H1(대립가설) : 자료가 정상성을 만족한다.

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(ts)

print("ADF Statistic : %f" % result[0])
print("p-value : %f" %result[1])
print("Critical Values")
for key, value in result[4].items():
    print("\t{} : {}".format(key, value))
```

##### 1차 차분

```python
ts.shift()
```

```python
ts.shift(2)
```

```python
ts_diff = ts-ts.shift()  # ts.shift() : 한칸 씩 밀리는 것. 괄호 안 숫자값을 입력하면 숫자값 만큼 뒤로 밀린다.
plt.figure(figsize = (12,7))
plt.plot(ts_diff)
plt.title("Differencing method")
plt.xlabel("Date")
plt.ylabel("Differencing Mean Temperature")
plt.show()
```

일정한 패턴이 확인되지 않고, 정상성을 만족하는 듯 보인다. 다시 ADF 검정을 진행한다.

```python
ts_diff
```

```python
result = adfuller(ts_diff[1:])
print("ADF Statistic : {}".format(result[0]))
print("p-value : {}".format(result[1]))
print("Critical Values :")
for key, value in result[4].items():
    print("\t{} : {}".format(key, value))
```

p-value가 0.05 이하의 값을 가지므로 귀무가설을 기각하여 정상성을 만족한다. (즉, 1차 차분한 데이터는 정상성을 만족한다.)

### 모델링 : ARIMA 모델의 설정

정상성을 만족하는 차분된 데이터로 ACF와 PACF 그래프를 그려 ARIMA 모형의 p와 q를 결정한다.

```python
fig = plt.figure(figsize = (12, 8))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_diff[1:], lags = 20, ax = ax1)  ## lag=20 > 20까지 보여준다.

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_diff[1:], lags = 20, ax = ax2)
```

acf, pacf 모두 금방 0에 수렴하고, acf의 경우 두번째 lag 이후 0에 수렴한다.
즉, ARIMA(2,1,2) 모형을 basemodel로, ARIMA(2,1,1), ARIMA(1,1,2), ARIMA(1,1,1)등의 모델을 시도해 볼 수 있다.

### 성능

```python
# R2, corr, mape, rmse로 성능 확인
from sklearn.metrics import r2_score, mean_absolute_percentage_error, mean_squared_error

def scoring(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    corr = np.corrcoef(y_true, y_pred)[0,1]
    mape = mean_absolute_percentage_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)

    print("r2 : {}\ncorr : {}\nmape : {}\nrmse : {}".format(r2, corr, mape, rmse))
```

```python
scoring(weather_bin.loc[weather_bin["Date"]>=start_index].MeanTemp, np.array(forecast))
```

r2및 corr가 0.9이상의 값을 가지므로 잘 예측됐음을 알 수 있다.

### 평가지표에 대해서

##### 정확도 
    - 정의 : 실제 데이터에서 예측 데이터가 얼마나 같은지 판단하는 지표
    - 특징 : 데이터 구성에 따라 머신러닝 모델의 성능을 왜곡할 가능성이 있음 -> 불균형한 레이블 값 지표의 경우 적합한 평가지표로 보기 어렵다.

##### 정밀도(Precision)
    - 정의 : Positive로 예측한 것들 중 실제로도 Positive인 것들의 비율
    - 특징 : Positive 예측 성능을 더욱 정밀하게 측정하기 위한 평가지표(양성 예측도라 불림)

##### 재현율(Recall)
    - 정의 : 실제 Positive인 것 중 Positive로 예측한 것들의 비율
    - 특징 : 민감도(Sensitivity) 혹은 TPR(True Positive Rate)이라고 불림

##### F1 스코어
    - 정의 : 정밀도와 재현율의 조화평균
    - 특징 : 정밀도와 재현율이 어느 한 쪽으로 치우치지 않고 적절한 조화를 이룰 때 상대적으로 높은 수치를 나타냄

##### AUC 스코어
    - 정의 : ROC 곡선 아래의 면적/ 1에 가까울수록 예측성능이 우수하다고 판단
    - 특징 : 가장 기본이 되는 값이 0.5(분류 능력이 없음)
    - roc_auc_score(y_test, y_score)로 y_score는 predict_proba()로 호출된 예측 확률 ndarray중 Positive 열에 해당하는 ndarray입니다.

```python
###LightGBM
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(n_estimators = 1000)
lgbm.fit(X_train, y_train)
```

```python
### LightGBM 예측/평가
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.metrics import f1_score, roc_auc_score
```

##### DNN

```python
import tensorflow as tf
```

```python
X_train.shape[1]
```

##### 피드포워드 신경망 구성

```python
# DNN 층 쌓기
## 정의
model = tf.keras.models.Sequential()

## Input Layer
model.add(tf.keras.layers.Dense(
    units = 64,
    input_dim = X_train.shape[1],
    activation = 'relu'))   ## 열의 수를 입력한다.

## Hidden Layer
model.add(tf.keras.layers.Dense(
    units = 32,
    input_dim = 64,
    activation = "relu"
))

## Output Layer
model.add(tf.keras.layers.Dense(
    units = 1,
    input_dim = 32,
    activation = "sigmoid"
))
```

```python
model.summary()
```

##### 피드포워드 신경망 훈련

```python
### 옵티마이저 설정
adam_optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001)

### 컴파일
model.compile(optimizer = adam_optimizer, loss = "binary_crossentropy", metrics = ["acc"])
```

```python
### 학습하기
history = model.fit(X_train, y_train, 
                    epochs = 15, 
                    batch_size = 64,
                    validation_split = 0.2)
```

```python
## 결과 확인
model.evaluate(X_test, y_test)
```

```python
y_pred_proba = model.predict(X_test,verbose = 0)
prd = y_pred_proba.argmax(axis =-1)
```

```python
prd_dnn = []
for i in y_pred_proba:
    if i>=0.5:
        prd_dnn.append(1)
    else:
        prd_dnn.append(0)
```

```python
prd_dnn
```

```python
## 평가
confusion = confusion_matrix(y_test, prd_dnn)
accuracy = accuracy_score(y_test, prd_dnn)
recall = recall_score(y_test, prd_dnn)
precision = precision_score(y_test, prd_dnn)
f1 = f1_score(y_test, prd_dnn)
auc = roc_auc_score(y_test, y_pred_proba)
```

```python
roc_auc_score(y_test, prd_dnn)
```

```python
## dnn 평가
print("오차행렬")
print(confusion)
print()
print(f"정확도 : {accuracy:.4f}, 재현율 : {recall:.4f}, 정밀도 : {precision:.4f}, f1 : {f1:.4f}, AUC : {auc :.4f}")
```

```python
from imblearn.over_sampling import SMOTE
```

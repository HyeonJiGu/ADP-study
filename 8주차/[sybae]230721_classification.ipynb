##### 데이터 확인(데이터 타입 및 결측치)

```python
# 결측치 확인하기
pd.set_option("display.max_rows",200)
pd.DataFrame(df.isnull().sum()*100/len(df))
```

```python
# 변수가 너무 많으므로 결측값이 존재하는 열과 결측 비율을 뽑아본다.
null_dict = {}
for var in df.columns:
    if df[var].isnull().sum()/len(df)>0:
        null_dict[var] = df[var].isnull().sum()/len(df)
```

```python
# "v2a1", "v18q1", "rez_esc"는 결측 비율이 70이상이므로 예측에 큰 영향을 주지 않아 제거하도록 하며, 나머지는 매우 비율이 작아 삭제하거나, 다른 값으로 대체한다.
null_dict
```

```python
# 데이터 정보 확인
df.info()
```

```python
# object인 값을 추출한다.(select_dtypes을 이용해 추출한다.)
df.select_dtypes(include=["object"])
```

Id, idhogar 모두 고유한 값을 갖는 변수이므로 삭제하도록 하며, dependency, edjefe, edjefa를 살펴보도록 한다.

```python
# dependency, edjefe, edjefa의 경우 yes = 1, no = 0으로 코드 설명서에 쓰여있다. 따라서 값을 대체해준다.
df.loc[df["edjefe"]=="yes","edjefe"]=1
df.loc[df["edjefe"]=="no","edjefe"]=0
```

```python
df.loc[df["edjefa"]=="yes","edjefa"]=1
df.loc[df["edjefa"]=="no","edjefa"]=0
```

```python
mapping = {"no" : 0, "yes" : 1}
df["dependency" ] = df["dependency"].replace(mapping)  ## 값을 수정할 땐 replace와 dictionary를 사용한다.
```

```python
df["edjefe"] = df["edjefe"].astype("float")
df["edjefa"] = df["edjefa"].astype("float")
df["dependency"] = df["dependency"].astype("float")
```

```python
# 삭제할 변수들 : "v2a1", "v18q1", "rez_esc" (결측치가 너무 많음), "Id", "idhogar" (고유한 값을 나타내는 컬럼)
remove_col = ["v2a1", "v18q1", "rez_esc", "Id", "idhogar"]
df_com = df.drop(remove_col, axis = 1).dropna()
```

```python
df_com.info()
```

```python
# 결측치가 있는지 확인하기 >> 없음
for var in df_com.columns:
    if df_com[var].isnull().sum()>0:
        print(var)
```

# Train/Test Split

### 이진 변수

1,2,3일 경우 취약계층에 속하지만 4는 취약계층에 속하지 않으므로 1,2,3은 0, 4는 1로 변환해 이진변수로 만든다.

```python
# 값 변환하기
binary = {1 : 0, 2:0, 3:0, 4:1}
df_com_new["binary"] = df_com_new["Target"].replace(binary)
```

```python
df_com_new["binary"] = df_com_new["Target"].replace(binary)
```

```python
X = df_com_new.drop(["binary", "Target"], axis = 1)
y = df_com_new["binary"]
```

```python
y.value_counts()/len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)
```

##### 선형회귀

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# 테스트 데이터셋의 성능
lr.score(X_train, y_train)
```

```python
pred_lr = lr.predict(X_test)
```

```python
# 혼동 행렬
cm = confusion_matrix(y_test, pred_lr)
print(cm)
```

```python
acc_lr = accuracy_score(y_test, pred_lr)
precision_lr = precision_score(y_test, pred_lr)
recall_lr = recall_score(y_test, pred_lr)
f1_lr = f1_score(y_test, pred_lr)
```

```python
# 로지스틱 회귀
print("acc : {}\nprecision : {}\nrecall : {}\nf1_score : {}".format(acc_lr, precision_lr, recall_lr, f1_lr))
```

```python
# ROC 커브 및 AUC 곡선
from sklearn.metrics import roc_curve, auc
fpr, tpr, thres = roc_curve(y_test, pred_lr, pos_label = 1)

# ROC 곡선 그리기
import matplotlib.pyplot as plt
plt.plot(fpr, tpr)
```

```python
# auc 스코어
auc = auc(fpr, tpr)
auc
```

##### Ensemble : Voting
- 하드 보팅 : 최종 아웃풋 결과 중 각 모델들이 가장많이 선택한 아웃풋을 최종 아웃풋으로 설정한다.
- 소프트 보팅 : 최종 아웃풋 결과의 확률값을 기반으로 평균을 내어, 이중 가장 확률값이 높은 아웃풋을 최종 아웃풋으로 설정한다.

```python
from sklearn.ensemble import VotingClassifier
### 소프트 보팅
voting_model = VotingClassifier(estimators =[("lr" , lr),
                                             ("mlp" , mlp),
                                             ("xgb" , xgb)], 
                                             voting = "soft")
```

```python
voting_model.fit(X_train, y_train)
```

```python
## Train 데이터 성능
voting_model.score(X_train, y_train)
```

```python
# 예측
# 오히려 과적합이 없는 것을 확인할 수 있다.
pred_vote = voting_model.predict(X_test)
print(accuracy_score(pred_vote, y_test))
```

**케라스**

```python
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
# target 값 원 핫 인코딩! > keras를 이용해 딥러닝을 진행할 때 y값은 원 핫 인코딩을 해줘야한다!(이진분류일 경우는 굳이 안해줘도 된다.)
N_class = 4
y_train_dnn = keras.utils.to_categorical(y_train-1, N_class)  ## -1을 한 이유는 인덱스가 0~3이 되어야 하는데 1~4가 되어 있기 때문에 맞춰주기 위해 -1을 해준다.
y_test_dnn = keras.utils.to_categorical(y_test-1, N_class)
```

```python
# 모델 만들기
model = keras.Sequential()

# 레이어 추가
model.add(keras.layers.Dense(units = 256, input_shape = (X_train.shape[1],)))  ## input_shape에는 데이터의 열의 개수를 넣어주면 된다.
model.add(keras.layers.Dense(128, activation = "relu"))
model.add(keras.layers.Dense(N_class, activation = "softmax"))
```

```python
model.summary()
```

```python
# 컴파일
model.compile(optimizer = "sgd"
              , loss = "categorical_crossentropy"
              , metrics = ["acc"])
```

```python
y_train.shape
```

```python
# 훈련하기
history = model.fit(X_train, y_train_dnn, epochs = 10)
```

```python
pred_dnn = model.predict(X_test)
pred_dnn
```

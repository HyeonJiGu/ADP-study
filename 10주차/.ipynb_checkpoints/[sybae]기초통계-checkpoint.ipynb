**7. 표본 10개의 분산이 90일 때 신뢰도 95%로 모분산의 신뢰구간을 추정**

```python
from scipy import stats
n = 10
rv = stats.chi2(df = n-1)
lcl = (n-1)*90/rv.isf(0.025)
hcl = (n-1)*90/rv.isf(0.975)
print("따라서 신뢰구간은 [{}.{}]이다.".format(lcl, hcl))
```

**9번. A생산라인의 제품 평균은 5.7mm이고 표준편차는 0.03, B생산라인의 제품 평균은 5.6mm이고 표준편차는 0.04라면 5%유의수준으로 두 제품의 평균이 차이가 있는지 여부를 검정하기 Z(0.05) = 1.65**

n의 값이 누락된 것으로 보임!(n값을 12로 가정하고 문제를 풀겠음)

- 귀무가설(H0) : 두 생산라인의 제품 평균은 차이가 없다.
- 대립가설(H1) : 두 생산라인의 제품 평균은 차이가 있다.

```python
ma = 5.7
mb = 5.6
sa = 0.03
sb = 0.04
na = 12
nb = 12
se = np.sqrt(((sa**2)/na) + ((sb**2)/nb))
z = (ma - mb)/se
t = stats.t(22)
p = 2*(1-t.cdf(z))
p 
```

```python
# GhatGPT 답
import numpy as np
from scipy.stats import t

# 주어진 정보
nA = 12
meanA = 5.7
stdA = 0.03

nB = 12
meanB = 5.6
stdB = 0.04

alpha = 0.05

# 두 그룹의 평균 차이와 표준오차 계산
mean_diff = meanA - meanB
std_error = np.sqrt((stdA**2 / nA) + (stdB**2 / nB))

# t-검정 통계량 계산
t_statistic = mean_diff / std_error

# 자유도 계산
degrees_freedom = nA + nB - 2

# p-값 계산
p_value = 2 * (1 - t.cdf(abs(t_statistic), df=degrees_freedom))

# 결과 출력
print("t-Statistic:", t_statistic)
print("p-Value:", p_value)

# 유의수준을 기준으로 검정 결과 결정
if p_value < alpha:
    print("귀무가설을 기각합니다. 두 제품의 평균은 차이가 있습니다.")
else:
    print("귀무가설을 기각하지 않습니다. 두 제품의 평균은 차이가 없습니다.")
```

**11.	주어진 데이터에서 신뢰구간을 구하려한다. 정규분포에서 표폰을 추출함**
- [Z(0.05) = -1.65 , Z(0.025) = -1.96, T(0.05, 8) = 1.860 , T0.025(0.025, 8) = 2.306]
- 데이터(9개) : [3.1, 3.3, 3.5, 3.7, 3.9, 4.1, 4.3 4.4, 4.7]
    - 1. 모분산을 모르는 경우 주어진 데이터의 95% 신뢰구간을 구하라
    - 2. sigma = 0.04인걸 알고 있을때의 95% 신뢰구간을 구하라


```python
# 1. 모분산을 모르는 경우
x = [3.1, 3.3, 3.5, 3.7, 3.9, 4.1, 4.3 ,4.4, 4.7]
# 준비할 것들
x_mean = np.mean(x)
n = len(x)
u_var = np.var(x, ddof = 1)
rv = stats.t(df= n-1)

# 구간 추정하기
lcl = x_mean - rv.isf(0.025)*np.sqrt(u_var/n)
ucl = x_mean - rv.isf(0.975)*np.sqrt(u_var/n)

print("모분산을 모르는 경우 95% 신뢰구간은 [{},{}]이다.".format(lcl, ucl))
```

```python
#2. 모분산을 아는 경우

# 준비할 것들
n = len(x)
rv = stats.norm()
x_mean = np.mean(x)
p_var = 0.04

# 구간 추정하기
lcl = x_mean-rv.isf(0.025)*np.sqrt(p_var/n)
ucl = x_mean-rv.isf(0.975)*np.sqrt(p_var/n)

print("모분산이 0.04인 경우 95% 신뢰구간은 [{},{}]이다.".format(lcl, ucl))
```

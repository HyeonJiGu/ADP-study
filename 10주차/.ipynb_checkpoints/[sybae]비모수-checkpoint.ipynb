# 적합도 검정

1. 상품 a와 b가 있을 때 다음과 같은 구매 패턴이 있다고 한다.
    - 1.1 구매 패턴으로 볼 때 두 상품이 연관이 있는지 가설을 세우고 검정하시오.
    - 1.2 가설을 채택하시오.

```python
x1 = "a a b b a a a a b b b b b a a b b a b b".split(" ")
x1
```

**1.1**
    - 귀무가설(H0) : 상품 a, b가 연관이 없다.
    - 대립가설(H1) : 상품 a, b가 연관이 있다.

**1.2**

```python
x1 = pd.Series(x1).replace({"a" : 0, "b" : 1}).to_list()
```

```python
from statsmodels.sandbox.stats.runs import Runs
import numpy as np
import pandas as pd

# Run 검정 분석(1,0으로 꼭 바꿔야 함!)
Runs(np.array(x1)).runs_test()
```

pvalue가 약 0.27로 유의수준 0.05보다 크다. 따라서 **귀무가설을 기각하지 않으며 상품 a,b가 연관이 없다고 결론내릴 수 있다.**

2. T여행사는 최근 개발한 북극탐험 여행 피키지 상품에 대한 만족도를 조사하기 위해 40명의 참가고객들을 대상으로 설문을 응답 받았다. 100점 만점으로 평가한 고객만족도 점수는 아래(x값)과 같다. T여행사는 다양한 마케팅 가설검증을 하기에 앞서 우선 이 고객 설문 결과가 기존의 고객 만족도 설문 결과와 같이 정규분포성을 띄는지 확인하려고 한다.
    - 2.1 귀무가설 및 대립가설을 작성하시오.
    - 2.2 검정통계량을 도출하고 가설 채택 여부를 기술하시오.
    - 2.3 만약 가정한 분포가 지수분포일 때의 검정통계량 및 pvalue를 도출하시오.

```python
x2 = [88,75,79,84,68,51,70,75,88,90,92,88,63,72,94,80,78,98,81,67,85,87,79,81,85,48,79,86,53,100,87,80,80,32,60,75,62,82,40,57]
```

**2.1**
    - 귀무가설(H0) : 고객 만족도 설문조사 점수는 정규분포를 따르지 않는다.
    - 대립가설(H1) : 고객 만족도 설문조사 점수는 정규분포를 따른다.

**2.2**

```python
from statsmodels.stats.diagnostic import kstest_normal

# Kolmogorov=Smirnov 검정
kstest_normal(x2, dist = "norm")
```

pvalue가 약 0.008로 유의수준 0.05보다 작으므로 귀무가설을 기각한다. 따라서 고객만족도 설문조사 점수는 정규분포를 따른다.

**2.3**

```python
print(kstest_normal(x2, dist= "exp"))
print("통계량은 {}이며, pvalue는  {}이다.".format(kstest_normal(x2, dist= "exp")[0], kstest_normal(x2, dist= "exp")[1]))
```

3. A초등학교 남학생 16명과 여학생 9명의 혈압을 측정한 pressure.csv파일을 가지고 남녀학생의 평균 혈압에 차이가 없는지 분석하라. 
    - 3.1 유의수준 0.1하에서 정규성 검정을 진행하시오.
    - 3.2 연구가설과 귀무가설을 설정하시오.
    - 3.3 통계량을 구하고, 유의수준 0.05하에서 연구가설 채택 여부를 기술하시오.
    - 3.4 만약 정규분포를 만족한다고 가정할 경우 유의수준 0.05하에서 등분산성을 검정하시오.(Bartlett, levene, fligner중 하나를 선택하고 그 이유를 작성하시오.)
    - 3.5 정규분포 및 등분산성을 만족할 경우 남학생과 여학생의 혈압차이에 대한 통계검정을 진행하시오.

```python
df3= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/26/pressure.csv')
df3.head(5)
```

**3.1**
- 귀무가설(H0) : 정규성을 충족한다.
- 대립가설(H1) : 정규성을 충족하지 않는다.

```python
from scipy.stats import shapiro
female = df3.loc[df3["gender"] == "female", "pressure"]
male = df3.loc[df3["gender"] == "male", "pressure"]

print("male의 정규성 검정 : ", shapiro(male))
print("female의 정규성 검정 : ", shapiro(female))
print()
print("male, female의 경우 pvalue가 0.1보다 작아 귀무가설을 기각하여 정규성을 만족하지 않는다는 결론이 나온다. 따라서 둘 중 하나라도 정규성을 만족하지 \
못하므로 비모수 검정을 진행한다.")
```

**3.2**
- 귀무가설(H0) : 남녀 학생 혈압의 차이가 없다.
- 대립가설(H1) : 남녀 학생 혈압의 차이가 있다.

**3.3**

```python
from scipy.stats import mannwhitneyu

# Mann-Whitney 검정 분석
print(mannwhitneyu(female, male))
print("pvalue가 0.14로 유의수준 0.05보다 크므로 귀무가설을 기각하지 않는다. 따라서 남녀 학생 혈압의 차이가 없다고 결론내릴 수 있다.")
```

**3.4**

- 귀무가설(H0) : 남, 녀의 분산이 동일하다.
- 대립가설(H1) : 남, 녀의 분산이 동일하지 않다.

```python
print("데이터가 정규분포인 경우에 등분산성 검정을 하기 위해선 Bartlett 검정을 진행해야 한다.")
print(stats.bartlett(female, male))
print("pvalue가 0.05보다 커 귀무가설을 기각하지 않는다. 따라서 두 집단의 분산이 동일하다.")
```

**3.5**

- 귀무가설(H0) : 남, 녀의 혈압은 동일하다.
- 대립가설(H1) : 남, 녀의 혈압은 동일하지 않다.

```python
print(stats.ttest_ind(female, male, equal_var = True))
print("pvalue가 0.12361716364314851이므로 0.05보다 커 귀무가설을 기각하지 않는다. 따라서 남, 녀 혈압은 동일하다.")
```

4. 강의 상류와 하류의 생물 다양성 점수에 차이가 있는지 검정하시오. (단, 같은 강에서 상류와 하류는 서로 독립적이지 않으며, 종속적인 관계에 있다./ 모든 검정은 유의수준 0.05하에서 진행된다.)
    - 4.1 정규성 검정을 진행하시오.
    - 4.2 귀무가설 및 연구가설을 설정하시오.
    - 4.3 **정규성을 만족하지 않는다고 가정할 때** 통계량을 구하고 연구가설 채택 여부를 기술하시오.

**4.1**

- 대립가설(H1) : 강의 상류와 하류의 생물 다양성 점수에 차이가 있다.

**4.3**

```python
from scipy.stats import wilcoxon
print(wilcoxon(df4["up"], df4["down"]))
print("pvalue가 0.24125289916992188로 0.05보다 크기 때문에 귀무가설을 기각하지 않는다. 따라서 강의 상류와 하류의 생물 다양성 점수에 차이가 없다.")
```

**4.4**

5. 공장 X,Y,Z의 평균 출하 소요시간을 여러 일자에 거쳐 측정한 데이터이다. 각 공장의 중위수의 차이가 존재하는지 확인하려 한다.(모든 검정은 유의수준 0.05하에서 진행된다.)
    - 5.1 정규성 검정을 진행하시오.
    - 5.2 귀무가설 및 연구가설을 작성하시오.
    - 5.3 **정규성을 만족하지 않는다고 가정할 때** 검정통계량을 구하고 연구가설 채택 여부를 기술하시오.
    - 5.4 각 공장별 소요시간이 정규성을 만족한다고 할 때 등분산성 검정을 진행하시오.(levene, bartlett, fligner 중 적절한 검정방법을 사용하시오.)
    - 5.5 각 공장별 소요시간이 정규성 및 등분산성 모두를 만족한다고 할 때 각 공장별 출하 소요시간에 차이가 있는지 귀무가설 및 연구가설을 설정하고 통계검정을 진행하시오.

```python
df5= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/25/problem5.csv',encoding='cp949')
df5.head(5)
```

**5.1**

```python
x = df5.loc[df5["공장명"] == "X","출하 소요시간"]
y = df5.loc[df5["공장명"] == "Y","출하 소요시간"]
z = df5.loc[df5["공장명"] == "Z","출하 소요시간"]
print("X공장의 정규성 검정 : ", shapiro(x))
print("Y공장의 정규성 검정 : ", shapiro(y))
print("Z공장의 정규성 검정 : ", shapiro(z))
print()
print("세 공장 모두 pvalue가 0.05보다 크므로 귀무가설을 기각하지 않는다. 따라서 모두 정규성을 충족한다.")
```

**5.2**
- 귀무가설(H0) : 각 공장의 중위수의 차이가 존재하지 않는다.
- 대립가설(H1) : 각 공장의 중위수의 차이가 존재한다.

**5.3**

```python
from scipy.stats import kruskal
print(kruskal(x,y,z))
print("pvalue가 0.05보다 크므로 귀무가설을 기각하지 않는다. 따라서 각 공장의 중위수의 차이가 존재하지 않는다.")
```

**5.4**

```python
print(stats.bartlett(x,y,z))
print("정규성이 만족될 경우 bartlett검정을 통해 등분산성을 검정한다. pvalue가 0.05보다 크므로 귀무가설을 기각하지 않는다. 따라서 분산이 동일하다.")
```

**5.5**

- 귀무가설(H0) : 공장별 출하 소요시간에는 차이가 없다.
- 대립가설(H1) : 공장별 출하 소요시간에는 차이가 있다.

```python
from scipy.stats import f_oneway
print(f_oneway(x,y,z))
print("정규성 및 등분산성이 모두 만족된 경우 f_oneway 검정을 진행한다.(정규성 만족 및 이분산인 경우welch_anova검정을 진행한다.) pvalue가 0.05보다 크므로 귀무가설을 기각하지 않는다. 따라서 공장별 출하 소요시간에 차이가 없다.")
```

# 상관성 검정

6. 타 홈쇼핑에 비해 뒤늦게 출발한 H 홈쇼핑에서는 최근 주력 판매 품목이었던 의류/패션 상품의 매출이 급락하여 시급한 대책마련이 필요하게 되었다. 패션을 포함한 소비재 상품을 취급하는 제 2상품본부장은 최근 의류 상품의 매출부진이 비교적 값 비싼 상품을 판매했기 때문이라고 지적하였고, CRM 팀에서는 이러한 지적이 사실인지 판단하기 위해 최근 판매된 5개 의류 브랜드의 가격과 판매량을 바탕으로 Kendall의 서열상관 계수를 구하기로 하였다. 아래의 x, y는 5개 브랜드에 대한 판매가격과 실제 판매량에 대해 서열척도로 변환한 것이다.(x6 = 판매가격 서열, y6 = 실제 판매량 서열)
    - 6.1 귀무가설 및 연구가설을 작성하시오.
    - 6.2 가격과 판매량에 대해 상관성이 있는지 검정통계량을 통해 귀무가설을 채택하고 결론을 도출하시오.

```python
x6 = [5,2,4,1,3]
y6 = [4,1,3,2,5]
```

**6.1**

- 귀무가설(H0) : 상품 가격과 판매량에는 상관관계가 없다.
- 대립가설(H1) : 상품 가격과 판매량에는 상관관계가 있다.

**6.2**

```python
from scipy.stats import kendalltau
kendalltau(x6, y6)
print("pvalue가 0.48로 0.05보다 커 귀무가설을 기각하지 않는다. 따라서 상관관계가 없다.")
```
